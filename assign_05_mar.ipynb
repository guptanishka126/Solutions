To create a Bokeh plot using python we have to follow these steps:
1) 'pip install bokeh' : Make sure you have installed the Bokeh library.
2) Import the necessary Bokeh modules: a- import bokeh.io, b- import bokeh.plotting, c- bokeh.io.output_notebook()
3) Import the necessary Bokeh class: from bokeh.plotting import figure, output_file,show
4) Prepare the data that you want to plot. This can be done by creating lists or arrays of values that represent the x-coordinates and y-coordinates of the data points.
5) Initialize a Bokeh figure object: p = figure()
6) Add glyphs (shapes) to the figure using the data you prepared: p.circle(x_values, y_values)

#Ans2. Glyphs are the building blocks of Bokeh visualizations. A glyph is a vectorized graphical shape or marker that is used to represent your data. Here are a few different categories of glyphs. They can be a markerâ€”shapes like circles, diamonds, squares, and triangles.

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

```python
#Ans2. 
#first of all import the required module and classes
from bokeh.plotting import figure,show,output_file
output_file("example1.html")
p = figure(title='Example of how to add a glyph in a Bokeh plot')
#give the values of x and y
x = [1,2,3,4,5]
y = [4,5,6,7,8]
#here, we are creating a scatter plot by adding a circle glyph to the figure using the circle() method.
p.circle(x,y)
#show output
show(p)
```

#Ans3. We can customize the appearance of a Bokeh plot, including the axes, title, and legend by setting the attributes on the figure object. Like p.title.text is used to customize the title of the Bokeh plot, p.xaxis.axis_label is used to set the x label of the X-axis while p.yaxis.axis_label is used to set the y label of the Y-axis and to add a legend label use legend_label attribute then pass any string to it.

```python
#Ans3. Example
from bokeh.plotting import figure,show,output_file
output_file("example2.html")
p = figure(title='Example of how to customize the plot appearance')
x = [1,2,3,4,5]
y = [4,5,6,7,8]
p.title.text = "Scatter plot"
p.xaxis.axis_label = "X - axis"
p.yaxis.axis_label = "Y - axis"

p.scatter(x,y, legend_label = "Blue points",size = 10)
#show output
show(p)
```

#Ans4. The Bokeh server is a component of the Bokeh library that allows you to create and serve interactive plots and applications that can be updated in real time. With the Bokeh server, you can create dynamic visualizations that respond to user interactions or change based on streaming data.

```python
#Example for Ans4. 
from bokeh.plotting import curdoc, figure
from bokeh.models import ColumnDataSource
from bokeh.layouts import column
from bokeh.events import Tap
initial_data = {'x': [1, 2, 3], 'y': [4, 5, 6]}
source = ColumnDataSource(initial_data)
def callback(event):
    # Modify the data in the ColumnDataSource object
    new_data = {'x': [4, 5, 6], 'y': [7, 8, 9]}
    source.data = new_data
p = figure()
p.circle('x', 'y', source=source)
p.on_event(Tap, callback)
layout = column(p)
curdoc().add_root(layout)
```
