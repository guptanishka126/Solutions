```python
#Ans1 .
import seaborn as sns
import plotly.express as px
titanic = sns.load_dataset('titanic')
fig = px.scatter(titanic, x='age', y='fare')
fig.update_layout()
fig.show()
```

```python
#Ans2 .
import plotly.express as px
import seaborn as sns
tips = sns.load_dataset("tips")
fig = px.box(tips,x ='day',y='total_bill')
fig.show()
```

```python
#Ans3.
import plotly.express as px
tips = px.data.tips()
fig = px.histogram(tips, x ='sex', y ='total_bill', pattern_shape='smoker', color='day')
fig.update_layout(xaxis_title="Sex", yaxis_title="Total bill", title="Histogram for Sex VS Total Bill")
fig.show()
```

```python
#Ans4 .
import plotly.express as px
iris = px.data.iris()
fig = px.scatter_matrix(iris,dimensions=['sepal_length','sepal_width','petal_length','petal_width'] , color='species')
fig.update_layout(title='Scatter matrix plot of iris dataset in Plotly')
fig.show()
```

#Ans5 .
The distplot figure factory displays a combination of statistical representations of numerical data, such as histogram, kernel density estimation or normal curve, and rug plot.

```python
#Ans 5 . Example
import plotly.figure_factory as ff
import plotly.express as px
iris = px.data.iris()
x = iris.petal_length
y = iris.sepal_length
hist_data = [x,y]
group_labels = ['Petal Length', 'Sepal Length']
fig = ff.create_distplot(hist_data, group_labels)
fig.show()
```
